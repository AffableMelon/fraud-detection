```python
# --- Section: Credit Card Data Analysis (Task 1 Full Credit) ---
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt

# Load the credit card data
df_credit = pd.read_csv('../data/raw/creditcard.csv')

# 1. Univariate Analysis: Class Distribution (Extreme Imbalance)
plt.figure(figsize=(6, 4))
sns.countplot(x='Class', data=df_credit, palette='magma', hue='Class')
plt.title('Bank Credit Card: Fraud (1) vs Legitimate (0)')
plt.show()
```

```python
# 2. Bivariate Analysis: Transaction Amount vs Class
plt.figure(figsize=(8, 5))
sns.boxplot(x='Class', y='Amount', data=df_credit)
plt.title('Transaction Amount by Class (Bank Data)')
plt.yscale('log') # Better visibility for outliers
plt.show()

```

```python
# 3. Statistical Summary
print("Fraud vs Legitimate Counts:")
print(df_credit['Class'].value_counts(normalize=True) * 100)
```
