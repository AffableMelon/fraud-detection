```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the newly created file
df = pd.read_csv('../data/processed/fraud_data_processed.csv')

# --- EDA TASK 1 REQUIREMENTS ---

# 1. Bivariate Analysis: Time Since Signup
# This usually shows a massive spike at 0-1 seconds for fraud cases
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='time_since_signup', hue='class', bins=50, element='step', common_norm=False)
plt.title('Fraud Probability based on Time Since Signup')
plt.show()
```

```python
# 2. Geolocation: Top Fraud Countries
fraud_rate = df.groupby('country')['class'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
fraud_rate.plot(kind='bar', color='crimson')
plt.title('Top 10 Countries by Fraud Rate')
plt.ylabel('Proportion of Transactions that are Fraud')
plt.show()

#Univariate Analysis
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Distribution of Age
sns.histplot(df['age'], bins=30, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of User Age')

# Distribution of Purchase Value
sns.histplot(df['purchase_value'], bins=30, kde=True, ax=axes[1], color='salmon')
axes[1].set_title('Distribution of Purchase Value')

plt.tight_layout()
plt.show()
```

```python
# 3. Class Imbalance Quantified
print(f"Fraud Class Percentage: {df['class'].mean() * 100:.2f}%")
```
